```python
uselog = pd.read_csv("use_log.csv")
uselog.head()
customer = pd.read_csv("customer_join.csv")
customer.head()
```

```python
customer_clustering = customer[["mean","max","min","membership_period"]]
customer_clustering
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
customer_clusterint_sc = sc.fit_transform(customer_clustering)

kmeans = KMeans(n_clusters=4,random_state=0)
clusters = kmeans.fit(customer_clusterint_sc)
customer_clustering = customer_clustering.assign(cluster = clusters.labels_)

print(customer_clustering["cluster"].unique())
customer_clustering
```
